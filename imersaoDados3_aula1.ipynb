```python
# Desafio 07

import matplotlib.pyplot as plt
# ordernar do menor para o maior valor
dados['tempo'].value_counts().sort_values().plot.bar()
plt.title('Experimentos por intervalo de tempo')
plt.ylabel('# de experimentos')
plt.xlabel('Tempo (horas)')
plt.show()
```

```python
drogas72D1 = dados.query('tempo == 72 and dose == "D1" and tratamento != "com_controle"')['droga'].unique().size
drogas72D2 = dados.query('tempo == 72 and dose == "D2" and tratamento != "com_controle"')['droga'].unique().size

print(f'''Drogas diferentes, aplicadas no experimento com intervalo de 72h, com dose D1: {drogas72D1}\nDrogas diferentes, aplicadas no experimento com intervalo de 72h, com dose D2: {drogas72D2}''')
```
